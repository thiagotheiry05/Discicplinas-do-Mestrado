As métricas de distância entre vetores são aplicadas nos estudos de aprendizagem de máquina como medidas de similaridade/dissimilaridade entre vetores que representam padrões. Apresente um estudo sobre as seguintes distancias:

- **Distancia Euclidiana**
- **Distancia de Minkowski**
- **Distancia de Mahalanobis**
- **Coeficiente de Correlação de Pearson**
- **Similaridade Cosseno**

Apresente neste estudo aplicações de métricas de distância em problemas de livre escolha.

*   Distância de Minkowski

A distancia de minkowski é uma generalização da distância euclidiana, parametrizada para medir diferentes tipos de distâncias entre dois vetores. Dada pela fórmula:

<div align="center">
<img src="https://i.stack.imgur.com/U6qGk.png"  width="300" height="100">
</div>

O parâmetro *p* controla qual tipo de distância é utilizada. Quando p = 1, a métrica de Minkowski é equivalente à distância de Manhattan, enquanto que quando p = 2, ela é equivalente à distância euclidiana.

A distancia de Minkowski pode ser aplicada em diferentes contextos onde se busca reconhecer padrões, como na analise de séries temporais para clusterização ou
processamento de imagens, quando o objetivo é medir a diferença entre duas regiões distintas numa imagem.

</br>

*   Distancia City Block

A fórmula da Distância City Block, também conhecida como Distância de Manhattan ou L1, é uma medida de distância entre dois pontos em um espaço de coordenadas com múltiplas dimensões.

<div align="center">
<img src="https://www.kdnuggets.com/wp-content/uploads/c_distance_metrics_euclidean_manhattan_minkowski_oh_5.png"  width="300" height="100">
</div>

Onde *n* é o número de dimensões do espaço e $x_1$ e $x_2$ são as coordenadas dos pontos nos eixos
i-ésimos.

Essa distância é útil em várias aplicações, como em análise de dados, reconhecimento de padrões e processamento de imagens, especialmente quando a diagonal não é uma opção de movimento e se deseja medir apenas a distância percorrida ao longo dos eixos.

*   Coeficiente de Correlação de Pearson

O coeficiente de correlação de Pearson é uma medida de similaridade que mede a relação linear entre duas variáveis. Ele varia de -1 a 1 e é dado pela fórmula:

<div align="center">
<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/602e9087d7a3c4de443b86c734d7434ae12890bc"  width="400" height="300">
</div>

Onde $x_i$ e $y_i$ são os valores das duas variáveis e $\hat{x}$ e $\hat{y}$ são suas médias.

* $\rho$ = 1 significa uma correlação perfeita positiva entre as duas variáveis.
* $\rho$ = -1 significa uma correlação negativa perfeita entre as duas variáveis - Isto é, se uma aumenta, a outra sempre diminui.
* $\rho$ = 0 significa que as duas variáveis não dependem linearmente uma da outra. No entanto, pode existir uma dependência não linear. Assim, o resultado
rho = 0 deve ser investigado por outros meios.

Na imagem em sequência podemos referenciar diferentes níveis de correlação:

<div align="center">
<img src="https://d33wubrfki0l68.cloudfront.net/e3b158e0207e1a80248bc0d7283f6eee638d5e72/4f6f0/blog/correlacao_files/figure-html/unnamed-chunk-4-1.png"  width="400" height="300">
</div>

O coeficiente de correlação de Pearson é usado em diversas aplicações, como análise de dados, análise financeira e reconhecimento de padrões. Por exemplo, em análise financeira, o coeficiente de correlação de Pearson é usado para medir a relação entre duas ações ou ativos financeiros e avaliar o risco de uma carteira de investimentos.

*   Similaridade Cosseno

A similaridade cosseno é uma medida de similaridade que mede o ângulo entre dois vetores no espaço vetorial. Ela é dada pela fórmula:

<div align="center">
<img src="https://miro.medium.com/v2/resize:fit:1400/0*0dJORNoFSi9uzeMf"  width="300" height="100">
</div>

Onde A . B representa o produto interno entre os vetores e ||A|| e ||B|| é o tamanho (norma) dos vetores.

<div align="center">
<img src="https://miro.medium.com/v2/resize:fit:640/0*f3cyZRboIFtX3NGl"  width="400" height="300">
</div>

A similaridade cosseno é comumente usada em sistemas de recomendação e recuperação de informações, onde a similaridade entre os vetores é usada para identificar itens semelhantes ou documentos relevantes para uma determinada consulta.

Por exemplo, em sistemas de recomendação de filmes, a similaridade cosseno pode ser usada para comparar as preferências de um usuário com as características de diferentes filmes e recomendar filmes que sejam mais relevantes para esse usuário.
